```python
import importlib
import numpy as np
import pickle
import tensorflow as tf
import random
import sys
import eli5
import d2d

import pandas as pd
import os
```

```python
file_high = "/data/fast0/datasets/Rhone_data_continuous_highpass.h5"
file_low = "/home/jmanos/notebooks/Lowpass_data/Rhone_data_continuous_lowpass.h5"

filt = 'Highpass'

input_columns = list(np.arange(0, 2308, 1))

seeds = list(np.arange(1,101,1))
bs = 32
ww = 200

history = {}
val_performance_high = {}
performance_high = {}

val_performance_low = {}
performance_low = {}

residuals_high = {}
true_discharge_vals_high = {}

residuals_low = {}
true_discharge_vals_low = {}

linear_model, lstm_model, dnn_model, df_all_chan_high, das_data_all, f  = d2d.import_data(filename = file_high, input_columns = input_columns)
linear_model, lstm_model, dnn_model, df_all_chan_low, das_data_all, f  = d2d.import_data(filename = file_low, input_columns = input_columns)

for SEED in seeds:
    multi_step_window_shuffled_high = d2d.WindowGenerator(df_all_chan_high,
                                                      input_width=ww,
                                                      label_width=1,
                                                      shift=0,
                                                      label_columns=['Discharge'],
                                                      input_columns=list(np.arange(0,2308,1)),
                                                      shuffle=True,
                                                      batch_size=bs,
                                                      seed = SEED)
    multi_step_window_shuffled_low = d2d.WindowGenerator(df_all_chan_low,
                                                  input_width=ww,
                                                  label_width=1,
                                                  shift=0,
                                                  label_columns=['Discharge'],
                                                  input_columns=list(np.arange(0,2308,1)),
                                                  shuffle=True,
                                                  batch_size=bs,
                                                  seed = SEED)

    lstm_model_high = tf.keras.models.load_model('saved_models/lstm_run'+str(SEED))
    val_performance_high['lstm_'+str(ww)+'_'+'high'+'_Fold:'+str(SEED)] = lstm_model_high.evaluate(multi_step_window_shuffled_high.val)
    performance_high['lstm_'+str(ww)+'_'+'high'+'_Fold:'+str(SEED)] = lstm_model_high.evaluate(multi_step_window_shuffled_high.test, verbose=0)
    
    lstm_model_low = tf.keras.models.load_model('saved_models/lstm_model_LF_'+str(SEED)+'.h5')
    val_performance_low['lstm_'+str(ww)+'_'+'low'+'_Fold:'+str(SEED)] = lstm_model_low.evaluate(multi_step_window_shuffled_low.val)
    performance_low['lstm_'+str(ww)+'_'+'low'+'_Fold:'+str(SEED)] = lstm_model_low.evaluate(multi_step_window_shuffled_low.test, verbose=0)
    
    ## DAS
    
    pred_on_test_high = lstm_model_high.predict(multi_step_window_shuffled_high.test).squeeze()
    pred_on_test_high = np.asarray(pred_on_test_high)

    predictions_test_high = (pred_on_test_high * multi_step_window_shuffled_high.dis_std) + multi_step_window_shuffled_high.dis_mean
    
    pred_on_test_low = lstm_model_low.predict(multi_step_window_shuffled_low.test).squeeze()
    pred_on_test_low = np.asarray(pred_on_test_low)

    predictions_test_low = (pred_on_test_low * multi_step_window_shuffled_low.dis_std) + multi_step_window_shuffled_low.dis_mean
    
    test_true_dis = []

    for i in multi_step_window_shuffled_high.test.as_numpy_iterator():
        test_true_dis.append(i[1])

    test_true_dis = (test_true_dis[0].squeeze() * multi_step_window_shuffled_high.dis_std) + multi_step_window_shuffled_high.dis_mean
    
    residuals_high['Run ' + str(SEED)] = predictions_test_high - test_true_dis
    residuals_low['Run ' + str(SEED)] = predictions_test_low - test_true_dis
    true_discharge_vals_high['Run ' + str(SEED)] = test_true_dis
    true_discharge_vals_low['Run ' + str(SEED)] = test_true_dis
```
