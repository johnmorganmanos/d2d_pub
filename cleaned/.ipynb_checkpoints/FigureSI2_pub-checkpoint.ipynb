```python
import numpy as np
import tensorflow as tf
import random
import sys
import eli5

import d2d
import importlib
importlib.reload(d2d)
import pandas as pd
import os
import d2d_all_data_window
importlib.reload(d2d_all_data_window)
```

```python

file = "/home/jmanos/notebooks/Highpass_50Hz_data/Rhone_data_continuous_highpass_50hz.h5"

filt = 'Highpass'

ww = 200
bs = 32



linear_model, lstm_model, dnn_model, df_all_chan, das_data_all, f  = d2d.import_data(filename = file)

da = df_all_chan.to_numpy()
```

```python
history = {}
val_performance = {}
performance = {}

residuals = {}
true_discharge_vals = {}

predictions = []

seeds = list(np.arange(1,101,1))
input_columns = [list(np.arange(0, 2308, 1)), list(np.arange(0, 1602, 1)), list(np.arange(1602, 2308, 1))]
location = ['Whole Cable', 'Ablation Zone', 'Accumulation Zone']



for SEED in seeds:
    
    whole = d2d_all_data_window.WindowGenerator(df_all_chan,
                                                  input_width=ww,
                                                  label_width=1,
                                                  shift=0,
                                                  label_columns=['Discharge'],
                                                  input_columns=list(np.arange(0,2308,1)),
                                                  shuffle=False,
                                                  batch_size=1)

    ablation = d2d_all_data_window.WindowGenerator(df_all_chan,
                                                  input_width=ww,
                                                  label_width=1,
                                                  shift=0,
                                                  label_columns=['Discharge'],
                                                  input_columns=list(np.arange(0, 1602, 1)),
                                                  shuffle=False,
                                                  batch_size=1)

    accumulation = d2d_all_data_window.WindowGenerator(df_all_chan,
                                                  input_width=ww,
                                                  label_width=1,
                                                  shift=0,
                                                  label_columns=['Discharge'],
                                                  input_columns=list(np.arange(1602, 2308, 1)),
                                                  shuffle=False,
                                                  batch_size=1)
    
    random.seed(SEED)
    np.random.seed(SEED)
    tf.random.set_seed(SEED)

    lstm_model = tf.keras.models.load_model('saved_models/lstm_model_highpass_50Hz_'+str(SEED)+'.h5')
    

    pred_on_test = lstm_model.predict(whole.train).squeeze()
    pred_on_test = np.asarray(pred_on_test)
    #pred_on_test = pred_on_test[:,-1]
    predictions_test = (pred_on_test * whole.dis_std) + whole.dis_mean
    
    predictions_test
    
    predictions.append(predictions_test)
    
    
    lstm_model = tf.keras.models.load_model('saved_models/lstm_run_'+str(SEED)+'_ablation')
    

    pred_on_test = lstm_model.predict(ablation.train).squeeze()
    pred_on_test = np.asarray(pred_on_test)
    #pred_on_test = pred_on_test[:,-1]
    predictions_test = (pred_on_test * ablation.dis_std) + ablation.dis_mean

    predictions.append(predictions_test)
    
    
    
    lstm_model = tf.keras.models.load_model('saved_models/lstm_run_'+str(SEED)+'_accumulation')
    

    pred_on_test = lstm_model.predict(accumulation.train).squeeze()
    pred_on_test = np.asarray(pred_on_test)
    #pred_on_test = pred_on_test[:,-1]
    predictions_test = (pred_on_test * accumulation.dis_std) + accumulation.dis_mean

    predictions.append(predictions_test)
    

        
```

```python
from sklearn.metrics import mean_absolute_error

mae_whole_100 = []
mae_abl_100 = []
mae_acc_100 = []

for i,j,k in zip(prediction_set[0::3],prediction_set[1::3],prediction_set[2::3]):

    whole_res_mae = mean_absolute_error(f['Discharge'][200::200], i)
    mae_whole_100.append(whole_res_mae)
    
    abl_res_mae = mean_absolute_error(f['Discharge'][200::200], j)
    mae_abl_100.append(abl_res_mae)
    
    acc_res_mae = mean_absolute_error(f['Discharge'][200::200], k)
    mae_acc_100.append(acc_res_mae)
```

```python
mae_whole_mean = np.mean(mae_whole_100)
mae_abl_mean = np.mean(mae_abl_100)
mae_acc_mean = np.mean(mae_acc_100)

mae_whole_std = np.std(mae_whole_100)
mae_abl_std = np.std(mae_abl_100)
mae_acc_std = np.std(mae_acc_100)
```
