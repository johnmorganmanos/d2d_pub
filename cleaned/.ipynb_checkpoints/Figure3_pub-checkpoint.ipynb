# PDD Model
#Using the GLAMOS data for glacier geometry

```python
import matplotlib.pyplot as plt
import matplotlib

melt_fact = 6 #mm/(day*degree C)
area_rhone = 15.5e12 #mm squared


lapse_rate_corr = 3 #celcius/1000m
ten_min_to_secs = 600 # seconds in 10 minuntes

mm3_to_m3 = 1e9

area_rhone_m = 15.5e6
glacier_length = 8e3 #units: meters
glacier_width = area_rhone_m / glacier_length

station_elevation = 1980 #units: meters

glacier_toe = 2200
glacier_head = 3600

temp = pd.read_pickle(r'temp_array.pkl')
dates_for_temp = pd.read_pickle(r'dates_4_temp.pkl')
precip =  pd.read_pickle(r'precip_mm.pkl')

times_for_dis = matplotlib.dates.date2num(f['Times'][:])
times = times_for_dis

melt_fact_cor = melt_fact / ( 1000 * 86400) #86400 seconds in a day, 1000 m in a mm
```

```python
elev_arr = np.arange(glacier_toe, glacier_head, 1)

corrected_elev = elev_arr - station_elevation
```

```python
PDD_dis = []
PDD_dis_precip = []

for i,k in zip(temp, precip):
    new_temp_arr = i-(lapse_rate_corr/1000)*corrected_elev
    for j in new_temp_arr:
        if j > 0:
            dis =   j * melt_fact_cor *  glacier_width*(np.sqrt(1+(glacier_length/1400)**2))
            dis_precip =  j * melt_fact_cor *  glacier_width*(np.sqrt(1+(glacier_length/1400)**2)) + (k* 0.001 * ((np.sqrt(1+(glacier_length/1400)**2)) * glacier_width)/ten_min_to_secs)
            PDD_dis.append(dis)
            PDD_dis_precip.append(dis_precip)
        else:
            dis = 0
            dis_precip = k* 0.001 * ((np.sqrt(1+(glacier_length/1400)**2)) * glacier_width)/ten_min_to_secs
            PDD_dis.append(dis)
            PDD_dis_precip.append(dis_precip)
```

```python
PDD_dis = np.asarray(PDD_dis)
PDD_dis = np.reshape(PDD_dis, (len(temp), len(corrected_elev)))
PDD_dis = np.sum(PDD_dis, axis=1)

PDD_dis_precip = np.asarray(PDD_dis_precip)
PDD_dis_precip = np.reshape(PDD_dis_precip, (len(temp), len(corrected_elev)))
PDD_dis_precip = np.sum(PDD_dis_precip, axis=1)
```

```python
#Cross-correlate to find the best time

import scipy.signal as signal

#PDD_interp = np.interp(times, dates_for_temp, PDD_dis)
discharge_interped = np.interp(dates_for_temp, times, f['Discharge'][:])
sig = discharge_interped
sig_noise = PDD_dis
corr = signal.correlate(discharge_interped, PDD_dis)
lags = signal.correlation_lags(len(PDD_dis), len(discharge_interped))
corr /= np.max(corr)
```

```python
fig, (ax_orig, ax_noise, ax_corr) = plt.subplots(3, 1, figsize=(4.8, 4.8))
ax_orig.plot(sig)
ax_orig.set_title('Original signal')
ax_orig.set_xlabel('Sample Number')
ax_noise.plot(sig_noise)
ax_noise.set_title('Signal with noise')
ax_noise.set_xlabel('Sample Number')
ax_corr.plot(lags, corr)
ax_corr.set_title('Cross-correlated signal')
ax_corr.set_xlabel('Lag')
ax_orig.margins(0, 0.1)
ax_noise.margins(0, 0.1)
ax_corr.margins(0, 0.1)
fig.tight_layout()
plt.show()
```

```python
file = "/home/jmanos/notebooks/Highpass_50Hz_data/Rhone_data_continuous_highpass_50hz.h5"

filt = 'Highpass'

input_columns = list(np.arange(0, 2308, 1))

seeds = list(np.arange(1,101,1))
bs = 32
ww = 200

history = {}
val_performance_DAS = {}
performance_DAS = {}

val_performance_TP = {}
performance_TP = {}

residuals_DAS = {}
true_discharge_vals_DAS = {}

residuals_TP = {}
true_discharge_vals_TP = {}

linear_model_DAS, linear_model_tp, lstm_model_DAS, lstm_model_tp, dnn_model_DAS, df_all_chan_DAS, input_columns_DAS, input_columns_tp, das_data_all, f  = d2d_tp.import_data(filename = file)

linear_model, lstm_model, dnn_model, df_all_chan, das_data_all, f  = d2d.import_data(filename = file, input_columns = input_columns)

for SEED in seeds:
    multi_step_window_shuffled = d2d.WindowGenerator(df_all_chan,
                                                      input_width=ww,
                                                      label_width=1,
                                                      shift=0,
                                                      label_columns=['Discharge'],
                                                      input_columns=list(np.arange(0,2308,1)),
                                                      shuffle=True,
                                                      batch_size=bs,
                                                      seed = SEED)
    
    windows = d2d_tp.WindowGenerator(df_all_chan_DAS,
                                     input_width=ww,
                                     label_width=1, 
                                     shift=0,
                                     label_columns=['Discharge'],
                                     input_columns_DAS=input_columns_DAS,
                                     input_columns_tp=input_columns_tp,
                                     shuffle=True,
                                     batch_size=16,
                                    seed = SEED)
        
    lstm_model = tf.keras.models.load_model('saved_models/lstm_model_highpass_50Hz_'+str(SEED)+'.h5')
    val_performance_DAS['lstm_'+str(ww)+'_'+filt+'_Fold:'+str(SEED)] = lstm_model.evaluate(multi_step_window_shuffled.val)
    performance_DAS['lstm_'+str(ww)+'_'+filt+'_Fold:'+str(SEED)] = lstm_model.evaluate(multi_step_window_shuffled.test, verbose=0)
    
    lstm_model_tp = tf.keras.models.load_model('saved_models/lstm_run_TP'+str(SEED))
    val_performance_TP['lstm_'+str(ww)+'_'+filt+'_Fold:'+str(SEED)] = lstm_model_tp.evaluate(windows.val_tp)
    performance_TP['lstm_'+str(ww)+'_'+filt+'_Fold:'+str(SEED)] = lstm_model_tp.evaluate(windows.test_tp, verbose=0)
    
    ## DAS
    
    pred_on_test = lstm_model.predict(multi_step_window_shuffled.test).squeeze()
    pred_on_test = np.asarray(pred_on_test)

    predictions_test = (pred_on_test * multi_step_window_shuffled.dis_std) + multi_step_window_shuffled.dis_mean
    
    test_true_dis = []

    for i in multi_step_window_shuffled.test.as_numpy_iterator():
        test_true_dis.append(i[1])

    test_true_dis = (test_true_dis[0].squeeze() * multi_step_window_shuffled.dis_std) + multi_step_window_shuffled.dis_mean
    
    residuals_DAS['Run ' + str(SEED)] = predictions_test - test_true_dis
    true_discharge_vals_DAS['Run ' + str(SEED)] = test_true_dis
    
    ## TP
    
    pred_on_test = lstm_model_tp.predict(windows.test_tp).squeeze()
    pred_on_test = np.asarray(pred_on_test)

    predictions_test = (pred_on_test * windows.dis_std_tp) + windows.dis_mean_tp
    
    test_true_dis = []

    for i in windows.test_tp.as_numpy_iterator():
        test_true_dis.append(i[1])

    test_true_dis = (test_true_dis[0].squeeze() * windows.dis_std_tp) + windows.dis_mean_tp
    
    residuals_TP['Run ' + str(SEED)] = predictions_test - test_true_dis
    true_discharge_vals_TP['Run ' + str(SEED)] = test_true_dis
    
    
    
    
    
    
melt_fact = 6 #mm/(day*degree C)
area_rhone = 15.5e12 #mm squared


lapse_rate_corr = 3 #celcius/1000m
ten_min_to_secs = 600 # seconds in 10 minuntes

mm3_to_m3 = 1e9

area_rhone_m = 15.5e6
glacier_length = 8e3 #units: meters
glacier_width = area_rhone_m / glacier_length

station_elevation = 1980 #units: meters

glacier_toe = 2200
glacier_head = 3600

temp = pd.read_pickle(r'temp_array.pkl')
dates_for_temp = pd.read_pickle(r'dates_4_temp.pkl')
precip =  pd.read_pickle(r'precip_mm.pkl')

```

```python
res_TP_arr = []
res_DAS_arr = []

true_TP_arr = []
true_DAS_arr = []

for i in residuals_TP.values():
    res_TP_arr.append(i)
    
for i in residuals_DAS.values():
    res_DAS_arr.append(i)
    
for i in true_discharge_vals_TP.values():
    true_TP_arr.append(i)
    
for i in true_discharge_vals_DAS.values():
    true_DAS_arr.append(i)
    
res_TP_arr = np.asarray(res_TP_arr)
res_DAS_arr = np.asarray(res_DAS_arr)

true_TP_arr = np.asarray(true_TP_arr)
true_DAS_arr = np.asarray(true_DAS_arr)


# Reshaping

res_TP_arr = np.asarray(res_TP_arr).reshape(res_TP_arr.shape[0] * res_TP_arr.shape[1])
res_DAS_arr = np.asarray(res_DAS_arr).reshape(res_DAS_arr.shape[0] * res_DAS_arr.shape[1])

true_TP_arr = np.asarray(true_TP_arr).reshape(true_TP_arr.shape[0] * true_TP_arr.shape[1])
true_DAS_arr = np.asarray(true_DAS_arr).reshape(true_DAS_arr.shape[0] * true_DAS_arr.shape[1])

DAS_mean = np.sum(np.abs(res_DAS_arr))
TP_mean = np.sum(np.abs(res_TP_arr))
PDD_mean = np.sum(np.abs(PDD_residual)) 

DAS_mae = np.around(np.mean(DAS_mean) / res_DAS_arr.shape[0], decimals = 2)
TP_mae = np.around(np.mean(TP_mean) / res_TP_arr.shape[0], decimals = 2)
PDD_mae = np.around(np.mean(PDD_mean) / PDD_residual.shape[0], decimals =2)

from scipy.stats import gaussian_kde

xy = np.vstack([true_TP_arr,res_TP_arr])
x1y1 = np.vstack([true_DAS_arr,res_DAS_arr])


z = gaussian_kde(xy)(xy)
z1 = gaussian_kde(x1y1)(x1y1)

x2y2 = np.vstack([f['Discharge'][shift:],PDD_residual])
z2 = gaussian_kde(x2y2)(x2y2)
```
