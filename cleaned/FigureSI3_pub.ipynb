```python
import os
os.environ["TF_ENABLE_ONEDNN_OPTS"]="0"
import numpy as np
import tensorflow as tf
import random

SEED = 1
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

import d2d
import importlib
importlib.reload(d2d)
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib

file = "/data/fast0/datasets/Rhone_data_continuous_highpass.h5"

filt = 'Highpass'

linear_model, lstm_model, dnn_model, df_all_chan, das_data_all, f  = d2d.import_data(filename = file)

multi_step_window_shuffled = d2d.WindowGenerator(df_all_chan,
                                                input_width=200,
                                                label_width=1, 
                                                shift=0,
                                                label_columns=['Discharge'],
                                                input_columns=list(np.arange(0,2308,1)),
                                                shuffle=True,
                                                batch_size=32)

normed_das = (das_data_all[:,0:2308] - multi_step_window_shuffled.chan_mean)/multi_step_window_shuffled.chan_std
```

```python
times = matplotlib.dates.date2num(f['Times'][:])
```

```python
files_per_day = int(60*60*24 / 30)
day_num = 21
day_end = day_num + 1
daily_rms = []
daily_var = []
times_for_rms = []
for i in range(22):
    day_start = times[i*files_per_day]
    times_for_rms.append(day_start)
    day_num = i
    day_end = day_num + 1
    mean_chan_rms = np.mean(normed_das[day_num*files_per_day:day_end*files_per_day,:], axis=0)
    std_chan_rms = np.std(normed_das[day_num*files_per_day:day_end*files_per_day,:], axis=0)
    daily_rms.append(mean_chan_rms)
    daily_var.append(std_chan_rms**2)
```

```python
temp = pd.read_pickle(r'temp_array.pkl')
dates_for_temp = pd.read_pickle(r'dates_4_temp.pkl')
precip =  pd.read_pickle(r'precip_mm.pkl')
wind = pd.read_pickle(r'wind_10minAVG.pkl')

precip_per_hour = [i * 6 for i in precip]

#Corrections
diff_in_alt = .9405 #km
lapse_rate_corr = 3.28 #celcius/1000m
new_temp = (np.asarray(temp) - (lapse_rate_corr)*diff_in_alt)
wind_per_hour = [i * 1 for i in wind]
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4,figsize=(14,5), gridspec_kw={'width_ratios': [6, 6, 1, 1]})
#fig.suptitle('DAS Deployment on Rhonegletscher, Switzerland', size = 20)
fig.patch.set_facecolor('white')

im1 = ax1.imshow(daily_rms, aspect = 'auto', cmap = 'seismic', vmin = -2, vmax =2, extent = [0, 2308,  times_for_rms[-1], times_for_rms[0]])
cbar1 = plt.colorbar(im1, ax=ax1)
ax1.grid()

im2 = ax2.imshow(daily_var, aspect = 'auto', cmap = 'Reds', vmin = 0, vmax=5,  extent = [0, 2308,  times_for_rms[-1], times_for_rms[0]])
cbar2 = plt.colorbar(im2, ax=ax2)
ax2.set_yticklabels([])
ax2.grid()

ax3.plot(new_temp, dates_for_temp, color='red', label='Temp.', lw=0.5)
ax3.yaxis_date()
ax3.set_ylim([times[0],times[-1]])
ax3.set_ylim(ax3.get_ylim()[::-1])
ax3.set_xlabel('Temp. ($^\circ$C)', size=12)
ax3.set_yticklabels([])
ax3.grid()

ax4.plot(wind_per_hour, dates_for_temp, color='green', label='Wind', lw=0.5)
ax4.set_ylim([times[0],times[-1]])
ax4.set_ylim(ax4.get_ylim()[::-1])
ax4.yaxis_date()
ax4.set_yticklabels([])
ax4.set_xlabel('Wind (m/s)', size=12)
#ax4.grid()

ax5 = ax3.twiny()
ax5.plot(precip_per_hour, dates_for_temp, color = 'blue', label = 'Precip.', lw = 0.5)
ax5.set_xlabel('precip (mm$/$h)')
ax3.grid()

cbar1.ax.set_ylabel('normalized DAS RMS strain rate', rotation=270, labelpad = -33)
cbar2.ax.set_ylabel('variance', rotation=270, labelpad = -16)

trans1 = ax1.get_xaxis_transform()
trans2 = ax2.get_xaxis_transform()
trans3 = ax3.get_xaxis_transform()
trans4 = ax4.get_xaxis_transform()


ax1.yaxis_date()
ax2.yaxis_date()

ax1.annotate('a. 24 hour mean 30 s RMS of DAS strain rate', weight = 'bold', xy=(0.02, 1.02), xytext=(0.02,  1.01), xycoords=trans1, annotation_clip=False)
ax2.annotate('b. 24 hour mean 30 s variance of DAS strain rate', weight = 'bold', xy=(0.02,  1.02), xytext=(0.02,  1.01),xycoords=trans2, annotation_clip=False)
ax3.annotate('c', weight = 'bold', xy=(0.02,  1.02), xytext=(17,  .96),xycoords=trans3, annotation_clip=False)
ax4.annotate('d', weight = 'bold', xy=(0.02,  1.02), xytext=(8.8,  .96),xycoords=trans4, annotation_clip=False)


ax1.set_ylabel('Date (YYYY-MM-DD)', size = 15)
ax1.set_xlabel('Channel', size = 15)
ax2.set_xlabel('Channel', size = 15)

ax5.spines['bottom'].set_color('red')
ax5.spines['top'].set_color('blue')

ax3.tick_params(axis='x', colors='red')
ax5.tick_params(axis='x', colors='blue')

#Add legend
# lines, labels = ax3.get_legend_handles_labels()
# lines2, labels2 = ax5.get_legend_handles_labels()
# ax3.legend(lines + lines2, labels + labels2, fontsize=7)

plt.tight_layout()
fig.savefig('RMS_and_variance_by_channel_meteo_compare.eps', format='eps')
plt.show()
```
